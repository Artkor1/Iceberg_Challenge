```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from os.path import join as opj
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pylab
plt.rcParams['figure.figsize'] = 10, 10
%matplotlib inline
```

```python
#Load the data
train = pd.read_json("train.json")
test = pd.read_json("test.json")
```

```python
print("train dataset: ", train.shape)
print("test dataset: ", test.shape)
```

```python
train
```

```python
X_band_1=np.array([np.array(band).astype(np.float32).reshape(75, 75) for band in train["band_1"]])
X_band_2=np.array([np.array(band).astype(np.float32).reshape(75, 75) for band in train["band_2"]])
X_train = np.concatenate([X_band_1[:, :, :, np.newaxis], X_band_2[:, :, :, np.newaxis],((X_band_1+X_band_2)/2)
                          [:, :, :, np.newaxis]], axis=-1)
```

```python
import plotly.offline as py
import plotly.graph_objs as go
py.init_notebook_mode(connected=True)
def plotmy3d(c, name):

    data = [
        go.Surface(
            z=c
        )
    ]
    layout = go.Layout(
        title=name,
        autosize=False,
        width=700,
        height=700,
        margin=dict(
            l=65,
            r=50,
            b=65,
            t=90
        )
    )
    fig = go.Figure(data=data, layout=layout)
    py.iplot(fig)
plotmy3d(X_band_2[12,:,:], 'iceberg')
```

```python
plotmy3d(X_band_1[14,:,:], 'Ship')
```

```python
target_train=train['is_iceberg']
x_train, x_test, y_train, y_test = train_test_split(X_train, target_train, random_state=1, train_size=0.75)
```

#1 Classification with Logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

logistic_train =  []
for log_train in x_train:
    logistic_train.append(log_train.flatten())

logistic_test = []
for log_test in x_test:
    logistic_test.append(log_test.flatten())

logistic_train = np.array(logistic_train)
logistic_test = np.array(logistic_test)
```

```python
logistic_model1 = LogisticRegression(verbose=1, max_iter=10, multi_class="multinomial", solver="sag")
logistic_model1.fit(logistic_train, y_train)
```

```python
logistic_model2 = LogisticRegression(verbose=1, max_iter=100, multi_class="multinomial", solver="sag")
logistic_model2.fit(logistic_train, y_train)
```

```python
logistic_model3 = LogisticRegression(verbose=1, max_iter=10, multi_class="multinomial", solver="saga")
logistic_model3.fit(logistic_train, y_train)
```

```python
logistic_model4 = LogisticRegression(verbose=1, max_iter=100, multi_class="multinomial", solver="saga")
logistic_model4.fit(logistic_train, y_train)
```

```python
acc_logistic1 = accuracy_score(y_test, logistic_model1.predict(logistic_test))
print("Logistic model #1 accuracy: ", acc_logistic1)

acc_logistic2 = accuracy_score(y_test, logistic_model2.predict(logistic_test))
print("Logistic model #2 accuracy: ", acc_logistic2)

acc_logistic3 = accuracy_score(y_test, logistic_model3.predict(logistic_test))
print("Logistic model #3 accuracy: ", acc_logistic3)

acc_logistic4 = accuracy_score(y_test, logistic_model4.predict(logistic_test))
print("Logistic model #4 accuracy: ", acc_logistic4)
```

#2 Classification with Neural Network (MLP) using MLPClassifier()

```python
from sklearn.neural_network import MLPClassifier
from sklearn import preprocessing
from sklearn.metrics import accuracy_score

mlp_classifier_train =  []
for mlp_train in x_train:
    mlp_classifier_train.append(mlp_train.flatten())

mlp_classifier_test = []
for mlp_test in x_test:
    mlp_classifier_test.append(mlp_test.flatten())

mlp_classifier_train = np.array(mlp_classifier_train)
mlp_classifier_test = np.array(mlp_classifier_test)

mlp_classifier1 = MLPClassifier(hidden_layer_sizes=(200,100,50),random_state=1)

mlp_classifier1.fit(preprocessing.StandardScaler().fit_transform(mlp_classifier_train), y_train)
```

```python
acc_mlp1 = accuracy_score(y_test, mlp_classifier1.predict(preprocessing.StandardScaler().fit_transform(mlp_classifier_test)))
print("Neural network model #1 accuracy: ", acc_mlp1)
```

#3 Classification with Keras model

```python
from matplotlib import pyplot
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Dropout, Input, Flatten, Activation
from keras.optimizers import Adam
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
model1 = Sequential()

#Dense Layer
model1.add(Dense(512, input_shape=(75, 75, 3), activation="relu"))

model1.add(Flatten())

#Sigmoid Layer
model1.add(Dense(1, activation="sigmoid"))


opt=Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0)
model1.compile(loss='binary_crossentropy',
            optimizer=opt,
            metrics=['accuracy'])
model1.summary()
```

```python
model1.fit(x_train, y_train,
          batch_size=24,
          epochs=5,
          verbose=1,
          validation_data=(x_test, y_test)
          )
```

```python
loss_model1 = model1.evaluate(x_test, y_test, verbose=1)[0]
acc_model1 = model1.evaluate(x_test, y_test, verbose=1)[1]
print('Keras model #1 test loss:', loss_model1)
print('Keras model #1 test accuracy:', acc_model1)
```

```python
model2 = Sequential()

#Dense Layer 1
model2.add(Dense(512, input_shape=(75, 75, 3), activation="relu"))
model2.add(Dropout(0.2))

#Dense Layer 2
model2.add(Dense(256, activation="relu"))
model2.add(Dropout(0.2))

#Dense Layer 3
model2.add(Dense(512, activation="softmax"))
model2.add(Dropout(0.2))

#Dense Layer 4
model2.add(Dense(256, activation="tanh"))
model2.add(Dropout(0.2))

model2.add(Flatten())

#Sigmoid Layer
model2.add(Dense(1, activation="sigmoid"))


opt=Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0)
model2.compile(loss='binary_crossentropy',
            optimizer=opt,
            metrics=['accuracy'])
model2.summary()
```

```python
model2.fit(x_train, y_train,
          batch_size=24,
          epochs=10,
          verbose=1,
          validation_data=(x_test, y_test)
          )
```

```python
loss_model2 = model2.evaluate(x_test, y_test, verbose=1)[0]
acc_model2 = model2.evaluate(x_test, y_test, verbose=1)[1]
print('Keras model #2 test loss:', loss_model2)
print('Keras model #2 test accuracy:', acc_model2)
```

```python
model3=Sequential()

#Conv Layer 1
model3.add(Conv2D(5, kernel_size=(3, 3), strides = (1, 1), padding='same', input_shape=(75, 75, 3),activation='relu'))

#Conv Layer 2
model3.add(Conv2D(5, (3, 3), strides = (3, 3), activation='relu'))
model3.add(MaxPooling2D(pool_size=(3, 3)))

#Dense Layer 1
model3.add(Dense(50, activation='relu'))

model3.add(Flatten())

#Sigmoid Layer
model3.add(Dense(1, activation="sigmoid"))


opt=Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0)
model3.compile(loss='binary_crossentropy',
            optimizer=opt,
            metrics=['accuracy'])
model3.summary()
```

```python
model3.fit(x_train, y_train,
          batch_size=24,
          epochs=50,
          verbose=1,
          validation_data=(x_test, y_test)
          )
```

```python
loss_model3 = model3.evaluate(x_test, y_test, verbose=1)[0]
acc_model3 = model3.evaluate(x_test, y_test, verbose=1)[1]
print('Keras model #3 test loss:', loss_model3)
print('Keras model #3 test accuracy:', acc_model3)
```

```python
model4=Sequential()

#Conv Layer 1
model4.add(Conv2D(5, kernel_size=(3, 3), strides = (1, 1), padding='same', input_shape=(75, 75, 3),activation='relu'))
model4.add(MaxPooling2D(pool_size=(3, 3), strides=(2, 2)))
model4.add(Dropout(0.2))

#Conv Layer 2
model4.add(Conv2D(5, (3, 3), strides = (3, 3), activation='relu'))
model4.add(MaxPooling2D(pool_size=(3, 3), strides=(2, 2)))
model4.add(Dropout(0.2))

#Dense Layer 3
model4.add(Dense(256, activation="relu"))
model4.add(Dropout(0.2))

#Dense Layer 4
model4.add(Dense(512, activation="relu"))
model4.add(Dropout(0.2))

model4.add(Flatten())

#Sigmoid Layer
model4.add(Dense(1, activation="sigmoid"))


opt=Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0)
model4.compile(loss='binary_crossentropy',
            optimizer=opt,
            metrics=['accuracy'])
model4.summary()
```

```python
model4.fit(x_train, y_train,
          batch_size=24,
          epochs=50,
          verbose=1,
          validation_data=(x_test, y_test)
          )
```

```python
loss_model4 = model4.evaluate(x_test, y_test, verbose=1)[0]
acc_model4 = model4.evaluate(x_test, y_test, verbose=1)[1]
print('Keras model #4 test loss:', loss_model4)
print('Keras model #4 test accuracy:', acc_model4)
```

```python
model5=Sequential()

#Conv Layer 1
model5.add(Conv2D(64, kernel_size=(3, 3),activation='relu', input_shape=(75, 75, 3)))
model5.add(MaxPooling2D(pool_size=(3, 3), strides=(2, 2)))
model5.add(Dropout(0.2))

#Conv Layer 2
model5.add(Conv2D(128, kernel_size=(3, 3), activation='relu' ))
model5.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
model5.add(Dropout(0.2))

#Conv Layer 3
model5.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
model5.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
model5.add(Dropout(0.2))

#Conv Layer 4
model5.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model5.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
model5.add(Dropout(0.2))


model5.add(Flatten())

#Dense Layers
model5.add(Dense(512, activation="relu"))
model5.add(Dropout(0.2))

#Dense Layer 2
model5.add(Dense(256, activation="relu"))
model5.add(Dropout(0.2))

#Sigmoid Layer
model5.add(Dense(1, activation="sigmoid"))



opt=Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0)
model5.compile(loss='binary_crossentropy',
            optimizer=opt,
            metrics=['accuracy'])
model5.summary()
```

```python
model5.fit(x_train, y_train,
          batch_size=24,
          epochs=80,
          verbose=1,
          validation_data=(x_test, y_test)
          )
```

```python
loss_model5 = model5.evaluate(x_test, y_test, verbose=1)[0]
acc_model5 = model5.evaluate(x_test, y_test, verbose=1)[1]
print('Keras model #5 test loss:', loss_model5)
print('Keras model #5 test accuracy:', acc_model5)
```
